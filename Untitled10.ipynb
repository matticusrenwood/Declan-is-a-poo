```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import matplotlib.image as mpimg
from scipy.ndimage import rotate

# Reload resized images
skyline_icon = mpimg.imread(skyline_resized_path)
evora_icon = mpimg.imread(evora_resized_path)

# Set up figure
fig, ax = plt.subplots(figsize=(10, 8))

# Plot track
ax.plot(left_boundary_x, left_boundary_y, linestyle='-', color='black', linewidth=2)
ax.plot(right_boundary_x, right_boundary_y, linestyle='-', color='black', linewidth=2)
ax.plot(racing_line_x, racing_line_y, linestyle='-', color='red', linewidth=1)

# Initialize car markers (empty, will update later)
skyline_marker = ax.imshow(skyline_icon, extent=[0, 0, 0, 0], zorder=3)
evora_marker = ax.imshow(evora_icon, extent=[0, 0, 0, 0], zorder=3)

# Compute direction angles at each point
dx = np.gradient(racing_line_x)
dy = np.gradient(racing_line_y)
angles = np.arctan2(dy, dx) * 180 / np.pi  # Convert to degrees

# Number of frames
num_frames = len(racing_line_x)

# Animation update function
def update(frame):
    skyline_x, skyline_y = racing_line_x[frame], racing_line_y[frame]
    evora_x, evora_y = racing_line_x[(frame + 30) % num_frames], racing_line_y[(frame + 30) % num_frames]

    # Rotate car icons based on track angle
    skyline_rotated = rotate(skyline_icon, -angles[frame], reshape=True)
    evora_rotated = rotate(evora_icon, -angles[(frame + 30) % num_frames], reshape=True)

    # Update images
    skyline_marker.set_data(skyline_rotated)
    evora_marker.set_data(evora_rotated)

    # Update positions
    car_size = 5
    skyline_marker.set_extent([skyline_x - car_size, skyline_x + car_size,
                               skyline_y - car_size, skyline_y + car_size])
    evora_marker.set_extent([evora_x - car_size, evora_x + car_size,
                             evora_y - car_size, evora_y + car_size])

    return skyline_marker, evora_marker

# Create animation
ani = animation.FuncAnimation(fig, update, frames=num_frames, interval=50, blit=False)

plt.show()
```
