```python
# --- Step 1: Install Dependencies ---
!pip install fastf1 matplotlib numpy pandas

# --- Step 2: Import Required Libraries ---
import fastf1 as f1
import fastf1.plotting
import matplotlib.pyplot as plt

# Enable FastF1 caching (for faster access to data)
f1.Cache.enable_cache('cache')

# --- Step 3: Load an F1 Session at Silverstone ---
year = 2023  # Select the year
gp = "Great Britain"  # Silverstone GP
session_type = "Q"  # Qualifying session

# Load session data
session = f1.get_session(year, gp, session_type)
session.load()

# Select a reference driver for telemetry (e.g., Max Verstappen)
driver = "VER"
lap = session.laps.pick_driver(driver).pick_fastest()  # Get fastest lap of the driver

# Get telemetry data (X = GPS Longitude, Y = GPS Latitude)
telemetry = lap.get_car_data().add_distance()
track_data = lap.get_pos_data()

# --- Step 4: Plot the Silverstone Track Map ---
plt.figure(figsize=(8, 10))
plt.plot(track_data['X'], track_data['Y'], linestyle='-', color='black', linewidth=2, label="Racing Line")
plt.scatter(track_data['X'], track_data['Y'], color='red', s=2)

# Labels and title
plt.xlabel("Longitude (X)")
plt.ylabel("Latitude (Y)")
plt.title("Silverstone Circuit - F1 GPS Data")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load the existing racing line data (assuming it's in X, Y coordinates)
racing_line_x = track_data["X"].values
racing_line_y = track_data["Y"].values

# Define track width (meters) - Approximate Silverstone width is ~14m
track_width = 14 / 2  # Half-width on each side

# Compute perpendicular vectors to racing line to define track boundaries
dx = np.gradient(racing_line_x)
dy = np.gradient(racing_line_y)
lengths = np.sqrt(dx**2 + dy**2)
perp_x = -dy / lengths  # Perpendicular unit vector X
perp_y = dx / lengths  # Perpendicular unit vector Y

# Generate left and right boundaries
left_boundary_x = racing_line_x + (perp_x * track_width)
left_boundary_y = racing_line_y + (perp_y * track_width)
right_boundary_x = racing_line_x - (perp_x * track_width)
right_boundary_y = racing_line_y - (perp_y * track_width)

# --- Plot Full Track with Racing Line ---
plt.figure(figsize=(10, 8))

# Plot left and right boundaries
plt.plot(left_boundary_x, left_boundary_y, linestyle='-', color='black', linewidth=2, label="Track Boundary")
plt.plot(right_boundary_x, right_boundary_y, linestyle='-', color='black', linewidth=2)

# Plot racing line
plt.plot(racing_line_x, racing_line_y, linestyle='-', color='red', linewidth=2, label="Racing Line")

# Labels and title
plt.xlabel("Longitude (X)")
plt.ylabel("Latitude (Y)")
plt.title("Silverstone Circuit with Racing Line & Full Track Width")
plt.legend()
plt.grid(True)
plt.show()

```
